### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Ingreso total Tienda 1
ingreso_t1 = tienda['Precio'].sum()
# Ingreso total Tienda 2
ingreso_t2 = tienda2['Precio'].sum()
# Ingreso total Tienda 3
ingreso_t3 = tienda3['Precio'].sum()
# Ingreso total Tienda 4
ingreso_t4 = tienda4['Precio'].sum()
```

```python
tiendas_nom = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
```

```python
ingresos = [ingreso_t1, ingreso_t2, ingreso_t3, ingreso_t4]
```

```python
df_ingresos = pd.DataFrame({
    'Tienda': tiendas_nom,
    'Ingreso': ingresos
})
```

```python
df_ingresos
```

```python
df_ingresos.plot(kind="bar", figsize=(12,8), x="Tienda", y="Ingreso", color="green", xlabel="Tienda", ylabel="Ingreso", title="Ingresos por tienda")
```

Se observa que la Tienda 1 presenta el mayor ingreso total, mientras que la Tienda 4 el menor.


# 2. Ventas por categoría

```python
def ventas_por_categoria(tienda, nombre_tienda):
  # Agrupar y contar ventas
  vent_categ = tienda.groupby("Categoría del Producto").size()
  # Convertir a DataFrame
  vent_categ =  vent_categ.reset_index(name="Cantidad")

  # Ordenar de mayor a menor
  vent_categ = vent_categ.sort_values(by="Cantidad", ascending=False)

  # Graficar
  # vent_categ.plot(kind="pie", figsize=(10,6), x="Categoría del Producto", y="Cantidad", xlabel="Categoría del Producto", ylabel="Cantidad", title=f"Ventas por categoría - {nombre_tienda}")
  vent_categ.plot(
    kind="pie",
    y="Cantidad",
    figsize=(8,8),
    labels=vent_categ["Categoría del Producto"],
    autopct="%1.1f%%",
    title=f"Ventas por categoría - {nombre_tienda}"
)


  return vent_categ
```

```python
ventas_por_categoria(tienda, "Tienda 1")
```

Se observa que los Muebles y Electrónicos son los productos con más ventas, al igual que Juguetes y Electrodomésticos.

```python
ventas_por_categoria(tienda2, "Tienda 2")
```

Se observa que los Muebles y Electrónicos son los productos con más ventas, al igual que Juguetes, Electrodomésticos y Deportes y Diversión.

```python
ventas_por_categoria(tienda3, "Tienda 3")
```

Se observa que los Muebles y Electrónicos son los productos con más ventas, al igual que Juguetes y Electrodomésticos.

```python
ventas_por_categoria(tienda4, "Tienda 4")
```

Se observa que los Muebles y Electrónicos son los productos con más ventas, al igual que Juguetes y Electrodomésticos.

# 3. Calificación promedio de la tienda

Se observa que la Tienda 3 es la que tiene el mejor promedio de calificación de los cliente, mientras que la Tienda 1 el menor.

## **Introducción**
El propósito de este análisis fue apoyar al Sr. Juan en la toma de decisión sobre cuál de las cuatro tiendas de la cadena Alura Store conviene vender para iniciar un nuevo emprendimiento. Para ello, se analizaron distintas métricas relacionadas con el desempeño de cada tienda, tales como ingresos totales, categorías y productos más vendidos, calificaciones de los clientes y costos de envío promedio. El análisis se apoyó en visualizaciones que permitieron comparar de forma clara los resultados obtenidos.


**1. Los ingresos totales de la tienda**

Los análisis de este punto, mostraron que la Tienda 1 es la que genera mayores ingresos ($1,150,880,000), mientras que la Tienda 4 tiene los ingresos más bajos ($1,116,344,000). No obstante, contar con mayores ingresos no necesariamente implica un mejor desempeño general, ya que este indicador debe analizarse junto con otros factores, como la satisfacción del cliente y los costos asociados a la venta.
Cabe mencionar que la Tienda 2 y la Tienda 3 presentan ingresos intermedios muy similares, lo que sugiere un desempeño estable.

**2. Las categorías de productos más y menos vendidos**

Al analizar las ventas por categoría, se observa que en las cuatro tiendas las categorías con mayor número de ventas son muebles, electrónicos, juguetes, electrodomésticos y deportes y diversión, en ese orden.
A pesar de que las diferencias entre categorías no son tan marcadas entre las tiendas, siguiendo la comparación entre la Tienda 1 y la Tienda 4 que se realizó anteriormente, esta última presenta una mayor cantidad de ventas en varias de estas categorías, a pesar de generar menores ingresos totales, lo que sugiere diferencias en el precio promedio de los productos vendidos.


**3. Las calificaciones promedio de los clientes por tienda**

Las calificaciones de los clientes representan un indicador importante porque de esta manera ayuda a medir la satisfacción del cliente con el servicio y productos. Los resultados muestran que la Tienda 3 cuenta con la calificación promedio más alta (4.04), seguida muy de cerca por la Tienda 2 (4.03), lo que indica un buen nivel de satisfacción.
Por otro lado, la Tienda 1 presenta el promedio de calificación más bajo (3.97), lo cual, aunque no representa una diferencia extrema, sí puede indicar posibles áreas de mejora en la experiencia del cliente.


**4. Los productos más y menos vendidos**

El análisis de los productos más y menos vendidos revela una mayor diversidad entre tiendas, lo que indica que cada una atiende necesidades de mercado distintas.
Es importante considerar que un mayor número de ventas de un producto no siempre se traduce en mayores ingresos, ya que el precio de cada producto juega un papel fundamental. En algunos casos, productos con menor volumen de ventas pueden generar ingresos comparables o superiores debido a su mayor precio.

**5. El coste de envío promedio para cada tienda**

En cuanto al costo de envío promedio, la Tienda 1 es la que tiene su valor más elevado ($26,018.60), mientras que la Tienda 4 tiene uno de los costos más bajos ($23,459.45). La tienda 2 y 3 de igual manera quedan en un punto intermedio. Aunque el envío es pagado por el cliente, un costo elevado puede provocar que éste no tenga el mismo incentivo para comprar, lo que afectaría la percepción del servicio, influyendo negativamente en la satisfacción y en la intención de recompra.
